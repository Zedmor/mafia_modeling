```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import json
```

```python
# Load the data
with open('src/mafia_game/logs/training_progress.json', 'r') as f:
    data = [json.loads(line) for line in f]

# Convert to pandas DataFrame
df = pd.DataFrame(data)
df = df[10000:]
```

```python
df['loss_log'] = df['loss'].apply(lambda x: np.log(x) if x > 0 else 0)

# Calculate moving averages
df['loss_log_smooth'] = df['loss_log'].rolling(window=1000).mean()
df['reward_smooth'] = df['reward'].rolling(window=1000).mean()
df['epsilon_smooth'] = df['epsilon'].rolling(window=1000).mean()
```

```python
# Plot the data
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(df['loss_log_smooth'], label='Loss')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(df['reward_smooth'], label='Reward')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(df['epsilon_smooth'], label='Epsilon')
plt.legend()

plt.show()
```
